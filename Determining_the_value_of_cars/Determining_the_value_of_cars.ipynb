### Объявление функций для построения графиков

```python
def show_num_variable(df, column, target=None):
    '''
    Функция отображения гистограммы распределения
    и диаграммы размаха для определенного столбца датафрейма
    с учетом принадлежности данного столбца к разным значениям
    переменной target.
    
    Параметры:
    - df: pandas.DataFrame, входной датафрейм
    - column: str, столбец для анализа
    - target: str или None, столбец для группировки (по умолчанию None)
    '''
    sns.set()
    f, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Гистограмма
    axes[0].set_title(f'Гистограмма для {column}', fontsize=16)
    axes[0].set_ylabel('Количество', fontsize=14)
    if target:
        sns.histplot(data=df, bins=20, kde=True, ax=axes[0], hue=target, x=column)
    else:
        sns.histplot(data=df, bins=20, kde=True, ax=axes[0], x=column)
    
    # Диаграмма размаха
    axes[1].set_title(f'Диаграмма размаха для {column}', fontsize=16)
    if target:
        sns.boxplot(data=df, ax=axes[1], x=target, y=column)
    else:
        sns.boxplot(data=df, ax=axes[1], y=column, orient='v')
    axes[1].set_ylabel(column, fontsize=14)
    
    plt.tight_layout()
    plt.show()
```

```python
def show_cat_variable_by_target(df, column, title, target=None, rot=60):
    '''
    Функция отображения соотношения категориальных признаков
    в столбце датафрейма, разделенных по значениям целевого признака.
    Если target не передан, отображается только countplot для column.
    Добавляет проценты на график.
    
    :param df: DataFrame, датафрейм с данными
    :param column: str, название столбца, по которому строится график
    :param title: str, заголовок графика
    :param target: str, название столбца с целевым признаком (по умолчанию None)
    :param rot: int, угол поворота меток на оси X
    '''
    
    # Если target не указан, рисуем только countplot для одного признака
    if target is None:
        plt.figure(figsize=(16, 6))
        ax = sns.countplot(data=df, x=column)
        ax.set_title(f"{title}", fontsize=14)
        ax.set_xlabel(column, fontsize=12)
        ax.set_ylabel('Количество', fontsize=12)
        ax.tick_params(axis='x', rotation=rot)
        
        # Добавляем проценты на график
        total_count = len(df)
        for p in ax.patches:
            count = p.get_height()
            if count > 0:  # Добавляем проценты только если высота столбца > 0
                percentage = f'{100 * count / total_count:.1f}%'
                ax.annotate(percentage,
                            (p.get_x() + p.get_width() / 2., count),
                            ha='center', va='center', 
                            xytext=(0, 10), 
                            textcoords='offset points',
                            fontsize=12, color='black')
        
        plt.tight_layout()
        plt.show()
        return
    
    # Проверяем, что столбец target существует в DataFrame
    if target not in df.columns:
        raise ValueError(f"Столбец '{target}' не найден в DataFrame.")
    
    # Если target указан, строим графики для каждого уникального значения target
    unique_targets = df[target].unique()
    num_targets = len(unique_targets)
    
    # Определяем размер холста в зависимости от количества таргетов
    fig, axes = plt.subplots(nrows=num_targets, ncols=1, figsize=(12, 5 * num_targets), sharex=True)
    
    # Если уникальных значений больше одного, axes будет массивом
    # Если только одно значение, делаем axes списком для унификации
    if num_targets == 1:
        axes = [axes]
    
    # Создаем график для каждого уникального значения целевого признака
    for i, target_value in enumerate(unique_targets):
        # Создаем подмножество данных для текущего значения целевого признака
        subset = df[df[target] == target_value]
        
        # Получаем уникальные значения категориального признака для сортировки
        categories = subset[column].value_counts().index.tolist()
        
        # Создаем countplot для текущего значения целевого признака
        ax = sns.countplot(data=subset, x=column, ax=axes[i], order=categories)
        
        # Заголовок и настройка осей
        axes[i].set_title(f"{title}: {target} - {target_value}", fontsize=14)
        axes[i].set_xlabel(column if i == num_targets - 1 else "", fontsize=12)  # Подпись только для нижнего графика
        axes[i].set_ylabel('Количество', fontsize=12)
        axes[i].tick_params(axis='x', rotation=rot)
        
        # Вычисляем общее количество для текущей группы
        total_count = len(subset)
        
        # Добавляем проценты на столбцы
        for p in ax.patches:
            count = p.get_height()
            if count > 0:  # Добавляем проценты только если высота столбца > 0
                percentage = f'{100 * count / total_count:.1f}%'
                # Вычисляем координаты для аннотации (верхняя часть столбца)
                ax.annotate(percentage,
                            (p.get_x() + p.get_width() / 2., count),
                            ha='center', va='center', 
                            xytext=(0, 10), 
                            textcoords='offset points',
                            fontsize=12, color='black')
    
    # Улучшаем отображение
    plt.tight_layout(h_pad=2.0)  # Добавляем вертикальный отступ между графиками
    plt.show()
```

## Машинное обучение

### Подготовка данных

```python
# Подготовка данных
X = autos_ml.drop(['price'], axis=1)
y = autos_ml['price']

num_columns = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_columns = X.select_dtypes(include=['category']).columns.tolist()

# Проверка
display(num_columns)
display(cat_columns)

# Разделение на train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
)
```

### Модель линейной регрессии

Линейная регрессия не умеет автоматически обрабатывать категориальные признаки и масштабировать количесвенные данные, поэтому проделаем это самостоятельно внутри Pipeline

```python
num_columns = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_columns = X_train.select_dtypes(include=['category']).columns.tolist()

# Проверка
display(num_columns)
display(cat_columns)
```

```python
# Обработка числовых и категориальных признаков с импутацией
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="no_info")),
    ("onehot", OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_columns),
        ("cat", categorical_transformer, cat_columns),
    ]
)
```

```python
# Пайплайн
linear_pipe = Pipeline(
    [
        ("preprocessing", preprocessor),
        ("model", LinearRegression()),
    ]
)

# Измерение времени обучения
start_time = time.time()
linear_pipe.fit(X_train, y_train)
training_time_linear = time.time() - start_time

print(f"⏱ Время обучения модели: {training_time_linear:.3f} секунд")
```

```python
start = time.time()  
y_pred_linear = linear_pipe.predict(X_train)
pred_time_linear = time.time() - start

print(f"⏱ Время предсказания модели: {pred_time_linear:.3f} секунд")
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_scores = cross_val_score(
    linear_pipe, 
    X_train, 
    y_train, 
    cv=kf, 
    scoring='neg_root_mean_squared_error'
)
rmse_linear = -np.mean(cv_scores)
print(f"✅ Среднее RMSE (CV): {rmse_linear:.3f} евро")
```

### Модель DecisionTreeRegressor

```python
# Обработка числовых и категориальных признаков с импутацией
numeric_transformer_tree = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer_tree = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="no_info")),
    ("ord", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=np.nan))
])

preprocessor_tree = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_tree, num_columns),
        ("cat", categorical_transformer_tree, cat_columns),
    ]
)
```

```python
# Пайплайн для модели
tree_pipe = Pipeline(
    [
        ("preprocessing", preprocessor),
        ("model", DecisionTreeRegressor(random_state=RANDOM_STATE)),
    ]
)

# Сетка гиперпараметров для подбора
param_grid = {
    'model__max_depth': [5],
    'model__min_samples_split': [2],
    'model__min_samples_leaf': [2],
}

# GridSearchCV
grid_search_tree = GridSearchCV(
    tree_pipe, 
    param_grid=param_grid,
    cv=5,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1
)
```

```python
# Измерение времени обучения для дерева
start_time = time.time()
grid_search_tree.fit(X_train, y_train)

# Вывод результатов
print(f"Лучший результат RMSE: {-grid_search_tree.best_score_:.2f}")
print("Лучшие параметры:", grid_search_tree.best_params_)

training_time_tree = time.time() - start_time

print(f"⏱ Время обучения дерева: {training_time_tree:.3f} секунд")
```

```python
# Время предсказания
best_model = grid_search_tree.best_estimator_
start_time = time.time()
y_pred_tree = best_model.predict(X_train)
pred_time_tree = time.time() - start_time

print(f"⏱ Время предсказания модели: {pred_time_tree:.3f} секунд")
```

```python
rmse_tree = -grid_search_tree.best_score_
print(f"✅ Среднее RMSE (CV): {rmse_tree:.3f} евро")
```

### Модель CatBoost

Модель CatBoost умеет работать с пропусками в данных, поэтому заполнять их с помощью Impute не нужно

```python
cb_model = cb.CatBoostRegressor(verbose=100, random_seed=RANDOM_STATE, loss_function='RMSE', iterations=500)

start_time = time.time()
cb_model.fit(cb_train)
training_time_cb = time.time() - start_time

print(f"⏱ Время обучения модели CatBoost: {training_time_cb:.3f} секунд")
```

```python
start_time = time.time()
cb_pred = cb_model.predict(cb_train)
pred_time_cb = time.time() - start_time
print(f"⏱ Время предсказания модели CatBoost: {pred_time_cb:.3f} секунд")
```

```python
# Пулы для обучения и теста
cb_train = cb.Pool(data=X_train, label=y_train, cat_features=cat_columns)
cb_test = cb.Pool(data=X_test, label=y_test, cat_features=cat_columns)

# Параметры модели
params = {
    "loss_function": "RMSE",
    "iterations": 500,
    "random_seed": RANDOM_STATE,
    "verbose": False
}

# Кросс-валидация
cv_results = cb.cv(
    params=params,
    pool=cb_train,
    fold_count=5,
    shuffle=True,
    partition_random_seed=RANDOM_STATE,
    verbose=False,
    plot=False
)

# Последнее значение RMSE по кросс-валидации
rmse_cb_cv = cv_results['test-RMSE-mean'].iloc[-1]
print(f"✅ Среднее RMSE по кросс-валидации (CatBoost cv): {rmse_cb_cv:.3f} евро")

```

### Модель LightGBM

Модель LightGBM не может напрямую работать с признаками типа datetime, поэтому данные признаки нужно преобразовать

```python
X_train_lgm, X_test_lgm = X_train, X_test

X_train_lgm['last_seen_day'] = X_train_lgm['last_seen'].dt.day
X_train_lgm['last_seen_month'] = X_train_lgm['last_seen'].dt.month
X_train_lgm['last_seen_year'] = X_train_lgm['last_seen'].dt.year
X_train_lgm = X_train_lgm.drop(columns=['last_seen'])

X_test_lgm['last_seen_day'] = X_test_lgm['last_seen'].dt.day
X_test_lgm['last_seen_month'] = X_test_lgm['last_seen'].dt.month
X_test_lgm['last_seen_year'] = X_test_lgm['last_seen'].dt.year
X_test_lgm = X_test_lgm.drop(columns=['last_seen'])
```

```python
lgb_train = lgb.Dataset(X_train_lgm, label=y_train)

params = {
    "objective": "regression",
    "metric": "rmse",
    "verbose": -1,
    "seed": 7
}
```

```python
start_time = time.time()
lgb_model = lgb.train(
    params=params,
    train_set=lgb_train,
)
training_time_lgb = time.time() - start_time

print(f"⏱ Время обучения модели LightGBM: {training_time_lgb:.3f} секунд")
```

```python
# результат вероятности
start_time = time.time()
lgb_pred = lgb_model.predict(X_train_lgm)
pred_time_lgb = time.time() - start_time
print(f"⏱ Время предсказания модели LightGBM: {pred_time_lgb:.3f} секунд")
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_scores = cross_val_score(
    lgb.LGBMRegressor(**params),
    X_train_lgm,
    y_train,
    cv=kf,
    scoring='neg_root_mean_squared_error'
)
rmse_lgb = -np.mean(cv_scores)
print(f"✅ Среднее RMSE (CV): {rmse_lgb:.3f} евро")
```
